Show a pic and crop the face. Andeas Brehme
the man who scored the entire victory for Germany in the 1990 World Cup

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('brehme.jpg') #read image
plt.imshow(img[...,::-1]) #show image with matplotlib
```

```python
cropped = img[30:350, 190:410] #crop the pic
plt.imshow(cropped[...,::-1]) # show cropped pic in matplotlib
cv.imwrite('brehme_face.jpg', cropped) # save cropped image
```
